Q-2--> Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:

A) Series
B) Place
C) Date
D) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome()
driver.get('https://www.bcci.tv/')

# for fixtures
fix=driver.find_element(By.XPATH,'//*[@id="imw-international-men"]/a[2]').click()

series=[]
place=[]
date=[]
time=[]

# for series
start=0
end=5
for page in range(start,end):
    se=driver.find_elements(By.XPATH,'//*[@id="match-card"]/div[1]/div/div/h5')
    for i in se:
        series.append(i.text)
        
# for place
start=0
end=5
for page in range(start,end):
    pl=driver.find_elements(By.XPATH,'//*[@id="match-card"]/div[2]/div[1]/span[1]')
    for i in pl:
        place.append(i.text)
        
        
# for Date 
start=0
end=5
for page in range(start,end):
    da=driver.find_elements(By.XPATH,'//*[@id="match-card"]/div[1]/div/div/div[2]/div[1]')
    for i in da:
        date.append(i.text)
    
# for time
for page in range(start,end):
    ti=driver.find_elements(By.XPATH,'//*[@id="match-card"]/div[1]/div/div/div[2]/div[2]')
    for i in ti:
        time.append(i.text)
        
```

```python
print(len(series),len(place),len(date),len(time))
```

```python
# for Data Frame
import pandas as pd
df=pd.DataFrame({'Series Name':series,'Place':place,'Date':date,'Time':time})
df.head()
```

Q-4--> Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:

A) Repository title
B) Repository description
C) Contributors count
D) Language used 

Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
driver=webdriver.Chrome()
driver.get('https://github.com/search?q=trading&type=repositories')
```

```python
repository=[]
description=[]
count=[]
language=[]

# For Repository Title 
start=0
end=5
for page in range(start,end):
    rep=driver.find_elements(By.CLASS_NAME,'bBwPjs')
    for i in rep:
        repository.append(i.text)
        repository=repository[0:50]
    
# For Reopsitory Description
for page in range(start,end):
    des=driver.find_elements(By.CLASS_NAME,'kWPXhV')
    for i in des:
        description.append(i.text)
        description=description[0:50]
        
        
# for Contributors count
for page in range(start,end):
    co=driver.find_elements(By.CLASS_NAME,'gPDEWA')
    for i in co:
        count.append(i.text)
        count=count[0:50]
        

# for Language Use
for page in range(0,5):
    lan=driver.find_elements(By.CLASS_NAME,'gVaQzD') # Box-sc-g0xbh4-0 
    for i in lan:
        language.append(i.text)
        language=language[0:50]
```

```python
print(len(repository),len(description),len(count),len(language))
```

```python
df=pd.DataFrame({'Repository Name':repository,'Description':description,'Contributors count':count,'Language Used':count})
df
```

```python
name=[]
types=[]
task=[]
attribute=[]
instances=[]
no_attribute=[]
year=[]

# for Dataset name
na=driver.find_elements(By.CLASS_NAME,'font-semibold') 
for i in na:
    name.append(i.text)
    
name.pop(0),name.pop(0),name.pop()


# for Data Types
types=[]
typ=driver.find_elements(By.CLASS_NAME,'col-span-3')
for i in typ:
    types.append(i.text)
    types=types[0:10]
    
    
# for attribute
att=driver.find_elements(By.CLASS_NAME,'col-span-3')
for i in att:
    attribute.append(i.text)
    attribute=attribute[0:10]
```
